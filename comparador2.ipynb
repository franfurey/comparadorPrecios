```python
def compare_products(df1, df2):
    from fuzzywuzzy import fuzz

    # Crear un DataFrame para almacenar los resultados
    df_result = pd.DataFrame(columns=['name_df1','name_df2', 'price_df1', 'price_df2', 'similarity'])

    # Comparar cada fila de df1 con cada fila de df2
    for i, row1 in df1.iterrows():
        name1 = row1['name']
        price1 = row1['price']
        found = False
        
        for j, row2 in df2.iterrows():
            name2 = row2['name']
            price2 = row2['price']
            
            # Calcular la similitud entre los nombres de los productos
            similarity = fuzz.token_set_ratio(name1, name2)
            
            # Si la similitud es mayor al umbral definido, tomar como el mismo producto
            if similarity >= 90:
                df_result.loc[len(df_result)] = [name1, name2, price1, price2, similarity]
                found = True
                break
        
        if not found:
            df_result.loc[len(df_result)] = [name1, None, price1, None, None]
    
    # Agregar productos de df2 que no coinciden con df1
    for i, row2 in df2.iterrows():
        name2 = row2['name']
        price2 = row2['price']
        found = False
        
        for j, row_result in df_result.iterrows():
            if row_result['name_df2'] == name2:
                found = True
                break
        
        if not found:
            df_result.loc[len(df_result)] = [None, name2, None, price2, None]


    df_result.to_csv('./df/resultado.csv', index=False)
    
    return True

```

```python
import pandas as pd
df_canal = pd.read_csv('./df/productos_canal.csv')
df_furey = pd.read_csv('./df/productos_furey.csv')

df_result = compare_products(df_canal, df_furey)

# Guardar el DataFrame resultante en un archivo CSV
```
